```python
transactions=[]#empty list
for i in range(0,7501):#this for loop iterates vertically
    transactions.append([str(data.values[i,j]) for j in range(0,20)])#iterates horizontlly
    
```

```python
from apyori import apriori
rules=apriori(transactions=transactions,min_support=0.003,min_length=2,max_length=2)
```

```python
results=list(rules)
```

```python
results
```

```python
def inspect(result):
    lhs         = [tuple(result[2][0][0])[0] for result in result]
    rhs         = [tuple(result[2][0][1])[0] for result in result]
    supports    = [result[1] for result in result]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(result), columns = ['prod 1', 'prod 2','Support'])
```
